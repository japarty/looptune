```python
import pandas as pd
import matplotlib.pyplot as plt
import pygwalker as pyg
```

```python
df = pd.read_csv('output/training_logs.csv')

# df = df[]

df = df[df['Epoch']<=30]

grouped = df.groupby(['model', 'timestamp']).filter(lambda x: len(x)>=2).groupby(['model', 'timestamp'])

plt.figure(figsize=(30,10))
for name, group_df in grouped:
    plt.plot(group_df['Epoch'], group_df['Accuracy'], label=name)

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Lines for Each Group')
plt.legend()
plt.show()

```

```python
# Individual subplots
fig, axes = plt.subplots(nrows=len(grouped.groups), ncols=1, figsize=(50, 50))

for (group_name, group_df), ax in zip(grouped, axes):
    group_df['Epoch'].plot(ax=ax, title=f"Category: {group_name}")

plt.tight_layout()
plt.suptitle("Subplots")
plt.show()
```
